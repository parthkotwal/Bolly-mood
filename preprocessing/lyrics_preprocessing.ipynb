```python
import os
import pandas as pd
data_path = os.path.join("..", "data","lyrics.csv")
df = pd.read_csv(data_path)
df
```

```python
import re
def remove_prefix(lyrics):
    lyrics = re.sub(r"^\d+\s*Contributor.*?Lyrics", "", lyrics) 
    lyrics = lyrics.strip()
    return lyrics

s = df.iloc[0,2]
print(remove_prefix(s))
```

```python
from transliterate import translit
def standardize_text(lyrics:str):
    return translit(lyrics, 'hi', reversed=True)

s = df.iloc[0,2]
print(standardize_text(s))
```
